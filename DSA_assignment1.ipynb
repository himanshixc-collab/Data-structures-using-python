```python
import numpy as np
```

```python
marks = np.random.randint(1, 101, size=(3, 16, 10))

print("3D Marks Array:")
print(marks)
print("Shape :", marks.shape)
```

```python
year_avg = np.mean(marks, axis=(1, 2))

print("Year-wise Average:")
for i in range(len(year_avg)):
    print("  Year", i+1, ":", round(year_avg[i], 2))
```

```python
student_total = np.sum(marks, axis=(0, 2))

print("Student-wise Total:")
for i in range(len(student_total)):
    print("  Student", i+1, ":", student_total[i])
```

```python
subject_high = np.max(marks, axis=(0, 1))

print("Subject-wise Highest Score:")
for i in range(len(subject_high)):
    print("  Subject", i+1, ":", subject_high[i])
```

```python
print("Marks below 40 before:", np.sum(marks < 40))

marks[marks < 40] = 40

print("Marks below 40 after :", np.sum(marks < 40))
print("Minimum mark now     :", np.min(marks))
```

```python
sorted_marks = np.sort(marks, axis=2)
print("After  sorting -", sorted_marks)
```
